```python
viz_df = pd.read_csv("../resources/Cleaned_Accidents_Data.csv")
viz_df.head()
```

```python
# // // - zip code color of US

plt1 = sns.relplot(
    data=viz_df,
    x='Longitude',
    y='Latitude',
    legend=False)

plt1.savefig("state_zipcode_zone1.png")
```

```python
plt2 = sns.relplot(
    data=viz_df,
    x='Longitude',
    y='Latitude',
    hue='State',
    legend=False)

plt2.savefig("state_zipcode_zone_colored.png")
```

```python
# state comparison

state_count = (viz_df.groupby(['State']).count()['Country']).to_list()

states_unique = viz_df['State'].unique()

fig, ax = plt.subplots()

ax = plt.barh(
    y=states_unique,
    width=state_count,
    align='center')

plt.show()

fig.savefig("state_count_total.png")
```

```python
N_paths = 5
N_steps = 100

x1 = np.linspace(0, 10, 100)
y = 0.1 * (np.random.random((N_paths, N_steps)) - 0.5)
y1 = y.cumsum(1)

x2 = np.linspace(0, 10, 100)
y = 0.1 * (np.random.random((N_paths, N_steps)) - 0.5)
y2 = y.cumsum(1)

fig2, ax = plt.subplots()
labels = ["a", "b",]
l1 = ax.plot(x1, y1.T, lw=4, alpha=0.1, c='b', label='a')
l2 = ax.plot(x2, y2.T, lw=4, alpha=0.2, c='r', label='b')

line_collections = [l1,l2]
plugins.connect(fig, plugins.InteractiveLegendPlugin(line_collections, labels))

mpld3.display()
```
